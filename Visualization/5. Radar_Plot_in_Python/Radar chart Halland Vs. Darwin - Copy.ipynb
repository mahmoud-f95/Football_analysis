### Radar Chart 

**Radar Chart:** Radar Charts are used to compare two or more players or teams on various characteristics or stats

**Data:**
The data used here is data pulled from fbref.com for the top 5 leagues player stats during 2021/2022 season

#### Importing library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from soccerplots.radar_chart import Radar
```

#### Importing data

```python
players=pd.read_csv("halland.csv")
df=players
df
```

#### Prepare the Radar Chart Arguments

```python
# Creating the Parameters
params=list(df.columns)
params=params[1:]
params
```

```python
# Add ranges
ranges=[]

for i in params:
    a=min(df[params][i])
    a=a-(a*0.15)

    b=max(df[params][i])
    b=b+(0.15*b)

    ranges.append((a,b))


ranges
    
```

```python
# Add each player stats
a_values=df.iloc[0:1].values.tolist()
b_values=df.iloc[1:].values.tolist()
a_values=a_values[0][1:]
b_values=b_values[0][1:]

values=[a_values,b_values]
values
```

```python
## title values
title = dict(
    title_name='Erling Haaland',
    title_color="red",
    subtitle_name='Borussia Dortmund 2021/2022 ',      ## subtitle on left side
    subtitle_color='black',
    title_name_2='Darwin Nunez',
    title_color_2="blue",    ## title on right side
    subtitle_name_2='Benfica 2021/2022',    ## subtitle on right side
    subtitle_color_2='black',
    title_fontsize=18,             ## same fontsize for both title
    subtitle_fontsize=15         ## same fontsize for both subtitle
)
endnote="@mahmoud_mf95 \ndata via FBREF"
```

#### Plotting

```python
#fig, ax = plt.subplots(figsize=(20, 10))     #and add figax=(fig,ax) in the plot args
#fig.set_facecolor('#FFFFFF')
radar=Radar()
fig,ax=radar.plot_radar(background_color="#FFFFFF",ranges=ranges,params=params,values=values,radar_color=['red', 'blue'],
                        alphas=[0.5,0.5],title=title,endnote=endnote,compare=True)


plt.savefig("Haalana Vs. Darwin Radar.png")
```
